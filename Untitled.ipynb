```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
LandscapeImage=np.array(Image.open('landscape.jpg').convert('L').resize((400, 300)))
Face=np.array(Image.open('face.jpg').convert('L').resize((400, 300)))
City=np.array(Image.open('city.jpg').convert('L').resize((400, 300)))
```

```python
plt.imshow(LandscapeImage)
```

```python
plt.imshow(Face)
```

```python
plt.imshow(City)
```

```python

LandscapeImageMatrix=np.asarray(LandscapeImage)
FaceMatrix=np.asarray(Face)
CityMatrix=np.asarray(City)
```

```python
print(LandscapeImageMatrix)
print(FaceMatrix)
print(CityMatrix)
```

```python
U1,S1,Vt1=np.linalg.svd(LandscapeImageMatrix,full_matrices=False)
U2,S2,Vt2=np.linalg.svd(FaceMatrix,full_matrices=False)
U3,S3,Vt3=np.linalg.svd(CityMatrix,full_matrices=False)
```

```python
k = 1000

compressed_landscape = U1[:, :k] @ np.diag(S1[:k]) @ Vt1[:k, :]
compressed_face      = U2[:, :k] @ np.diag(S2[:k]) @ Vt2[:k, :]
compressed_city      = U3[:, :k] @ np.diag(S3[:k]) @ Vt3[:k, :]
```

```python
plt.imshow(compressed_landscape,cmap='gray')
plt.title(f"Reconstructed with k={k}")
plt.show()
```

```python
plt.imshow(compressed_face ,cmap='gray')
plt.title(f"Reconstructed with k={k}")
plt.show()
```

```python
plt.imshow(compressed_city,cmap='gray')
plt.title(f"Reconstructed with k={k}")
plt.show()
```
